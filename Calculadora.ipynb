## Vamos começar?
Para iniciar a calculadora, pressione o botão para rodar a célula de código seguinte.

```python
def calculadora():
    operacao = input('''
Cálculos matemáticos são difíceis. Deixa que eu faço para você. Qual operação você deseja realizar?
  Escreva + para adição,
  - para subtração,
  * para multiplicação,
  / para divisão,
  // para radiciação e
  ** para potenciação.
''')

    num_1 = int(input('Ok, entendi. Agora, o primeiro número da operação é: '))
    num_2 = int(input('E o segundo: '))

    if operacao == '+':
        print('{} + {} = '.format(num_1, num_2))
        print(num_1 + num_2)

    elif operacao == '-':
        print('{} - {} = '.format(num_1, num_2))
        print(num_1 - num_2)

    elif operacao == '*':
        print('{} * {} = '.format(num_1, num_2))
        print(num_1 * num_2)

    elif operacao == '/':
        print('{} / {} = '.format(num_1, num_2))
        print(num_1 / num_2)

    elif operacao == '**':
        print('{} elevado a {} = '.format(num_1, num_2))
        print(num_1 ** num_2)

    elif operacao == '//':
        print('A raiz de {} na base {} é =')
        print(num_1 // num_2)

    else:
        print('Não entendi. Poderia repetir o comando?')

    mais_um()

def mais_um():
    calcular_novamente = input('''
Essa foi fácil. Posso te ajudar de novo?
Escreva S para sim e N para não.
''')

    if calcular_novamente.upper() == 'S':
        calculadora()
    elif calcular_novamente.upper() == 'N':
        print('Tudo bem. Continuo à disposição. :)')
    else:
        print('Ops, não entendi.')
        mais_um()

calculadora()
```
