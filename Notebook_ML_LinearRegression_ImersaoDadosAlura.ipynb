Esses são os dados com que ficamos após a limpeza e a análise exploratória.

Considerando nosso objetivo final, que é a predição do valor do imóvel de acordo com a região de São Paulo, uma informação útil de extrair desse conjunto é o valor do m².

Sendo assim, vamos criar uma nova coluna em dados_vendas chamada **Valor_m²**, que consistirá na divisão do Valor_anuncio_float e Metragem.

```python
dados_vendas["Valor_m2"] = dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"]
dados_vendas.head()
```

Agora que conhecemos o valor do m² de cada imóvel, o próximo passo é obter um conjunto de dados que nos dê essa informação média por bairro.

Para atingir esse objetivo, vamos usar a função .groupby(), que pede uma coluna com chaves repetidas para agrupamento (Bairro) e, nesse caso, definimos que os valores numéricos serão agrupados em seu valor médio (e não a soma).

```python
dados_vendas.groupby("Bairro").mean()
```

Vamos armazenar esse dado em outra variável chamada dados_bairro. Entretanto, os instrutores do curso sugeriram que a média fosse calculada manualmente. Sendo assim, inicialmente armazenamos dados_vendas agrupado por bairro em soma em dados_bairro.

```python
dados_bairro = dados_vendas.groupby("Bairro").sum()
dados_bairro
```

Depois, atualizamos a variável, adicionando a coluna Valor_m2_bairro como a divisão do Valor_anuncio_float por Metragem. Chamando dados_bairro, vemos que obtivemos um resultado bem semelhante ao chamar a função groupy().mean()

```python
dados_bairro["Valor_m2_bairro"] = dados_bairro["Valor_anuncio_float"]/dados_bairro["Metragem"]
dados_bairro
```

Podemos reparar aqui, que usando groupby().mean() em dados_vendas, temos a função média preservada, e em dados_bairro, temos a soma dos dados. Em ambos, usamos a função .loc[] para localizar a linha do bairro Vila Mariana.

```python
dados_vendas.groupby("Bairro").mean().loc["Vila Mariana"]
```

```python
dados_bairro.loc["Vila Mariana"]
```

# Modelo de regressão linear

A regressão linear é considerada um tipo de aprendizado de máquina supervisionado. Com nossa correlação, criamos um eixo numérico pelo qual a máquina consegue predizer valores a partir de sua proximidade com o eixo. 

Esse tipo de algoritmo é muito utilizado para previsões do preço de ações, análise de demanda e volume de vendas, diagnóstico médico e quaisquer correlações numéricas, como o nosso caso.

Obtenha mais informações [aqui](https://www.datageeks.com.br/machine-learning/).

Tipos-de-Machine-Learning.webp

Começaremos importando os pacotes LinearRegression e train_test_split das bibliotecas do sklearn.

Estabelecemos X = Metragem e Y = Valor_anuncio.

train_test_split é uma função que gera novos subsets, evitando algum tipo de bias entre o treinamento e a testagem do modelo.

A função utiliza um valor bidimensional de X e um valor unidimensional de Y. Os outros parâmetros são opcionais:
  test_size é a % do dataset utilizada para os testes. No caso, nós estabelecemos que será 25%. O random_state é um int que controla a randomização durante a divisão dos subsets.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = dados_vendas_censo[["Metragem"]]
Y = dados_vendas_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54)
lr = LinearRegression()
```
